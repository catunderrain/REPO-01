```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data = pd.read_csv('a.csv').values
N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
plt.scatter(x, y)
plt.xlabel('mét vuông')
plt.ylabel('gía')
x = np.hstack((np.ones((N,1)),x))
w = np.array([0.,1.]).reshape(-1,1)
numOfIteration = 100
cost = np.zeros((numOfIteration,1))
learning_rate = 0.0001
for i in range(1, numOfIteration):
    r = np.dot(x, w) - y
    cost[i] = 0.5*np.sum(r*r)
    w[0] -= learning_rate*np.sum(r)
    # correct the shape dimension
    w[1] -= learning_rate*np.sum(np.multiply(r, x[:,1].reshape(-1,1)))
predict = np.dot(x, w)
plt.plot((x[0][1], x[N-1][1]),(predict[0], predict[N-1]),'k')
plt.show()
lrg = LinearRegression()
lrg.fit(x, y)
y_pred = lrg.predict(x)
plt.plot((x[0], x[-1]),(y_pred[0], y_pred[-1]), 'b')
plt.show()
np.savez('w2.npz', a=lrg.intercept_, b=lrg.coef_)
k = np.load('w2.npz')
lrg.intercept_ = k['a']
lrg.coef_ = k['b']
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
x = np.arange(-20,20,0.1)
y = 1/(1 + math.e**(-1*x))
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
x = np.arange(-20,20,0.1)
y = (math.e**x - math.e**(-1*x))/(math.e**x + math.e**(-1*x))
plt.plot(x,y)
plt.show()
```
