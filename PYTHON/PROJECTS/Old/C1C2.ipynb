```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 3 * np.pi, 0.0001)
y_sin = np.sin(np.tan(x))
y_cos = np.cos(np.tan(x))
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('Con bo duc')
plt.ylabel('Con ga trong')
plt.title('Dit me bo may chap het')
plt.legend(['Cac', 'Lon'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# T‰nh tọa đº x, y cho đồ thị sin, cos
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.xlabel('vai')
plt.ylabel('lon')
plt.legend(['Cac'])
plt.title('Sine')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.xlabel('vai')
plt.ylabel('câc')
plt.legend(['Lon'])
plt.title('Cosine')

plt.show()
```
